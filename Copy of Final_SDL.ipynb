## Importing Packages

```
import pandas as pd
import numpy as np
from sklearn import metrics
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
```

## Reading CSV files


```
dataset1=pd.read_csv('/content/drive/MyDrive/sdl_project/csv Files/Bitcoin Cash.csv')
dataset2=pd.read_csv('/content/drive/MyDrive/sdl_project/csv Files/Bitcoin SV US Dollar.csv')
dataset3=pd.read_csv('/content/drive/MyDrive/sdl_project/csv Files/Ethereum.csv')
dataset4=pd.read_csv('/content/drive/MyDrive/sdl_project/csv Files/Litecoin.csv')
```

# Bitcoin Cash

```
dataset1['DATE'] = pd.to_datetime(dataset1.DATE)
```

```
X1  = dataset1[['OPEN','HIGH','LOW']]
y1 = dataset1['PRICE']
```

```
X1_train , X1_test , y1_train , y1_test = train_test_split(X1,y1,random_state = 0)
regressor1 = LinearRegression()
regressor1.fit(X1_train,y1_train)
```

```
print(X1_train[:])
print(X1_test[:])
```

```
print(regressor1.coef_)
print(regressor1.intercept_)
```

```
predicted_btc=regressor1.predict(X1_test)
```

```
dframe=pd.DataFrame(y1_test,predicted_btc)
```

```
dfr=pd.DataFrame({'Actual':y1_test,'Predicted':predicted_btc})
```

```
regressor1.score(X1_test,y1_test)
```
